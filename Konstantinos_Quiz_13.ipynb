```
def total(items, prices):
    tot=0
    foods = ["bread", "cheese", "butter", "onigiri", "chicken", "tomato"]
    electronics = ["ipad", "iphone", "macbook", "phone", "laptop", "desktop"]
    alcohol = ["beer", "wine", "vodka", "rum", "sake"]
    for i, j in enumerate(items):

        if j in foods:
            tot += prices[i] * 1.1

        elif j in electronics:
            tot += prices[i] * 1.15

        elif j in alcohol:
            tot += prices[i] * 1.2

        else:
            return f"item '{j}' not in database, please add it. "
    return tot

```

```
print('total(["bread", "beer", "ipad"],[300, 800, 30000])', "==>", total(["bread", "beer", "ipad"],[300, 800, 30000]))
print('total(["apple", "ipad", "laptop"],[700, 33000, 33000])', "==>", total(["apple", "ipad", "laptop"],[700, 33000, 33000]))
print('total(["wine", "phone", "cheese", "tomato", "rum"],[700, 29000, 500, 250, 900])', "==>",
      total(["wine", "phone", "cheese", "tomato", "rum"],[700, 29000, 500, 250, 900]))
```
