```
def total(items, prices):
    tot=0

    for i, j in enumerate(items):

        if j == "bread":
            tot += prices[i] * 1.1

        elif j == "ipad":
            tot += prices[i] * 1.15

        else:
            tot += prices[i] * 1.2

    return tot
```

```
print('total(["bread", "beer", "ipad"],[300, 800, 30000])', "==>", total(["bread", "beer", "ipad"],[300, 800, 30000]))
print('total(["beer", "ipad", "ipad"],[700, 33000, 33000])', "==>", total(["beer", "ipad", "ipad"],[700, 33000, 33000]))
print('total(["beer", "ipad", "bread", "bread", "beer"],[700, 29000, 500, 250, 800])', "==>",
      total(["beer", "ipad", "bread", "bread", "beer"],[700, 29000, 500, 250, 800]))
```
