Quiz 17


```
class ABC():
    def __init__(self, a, b, c):
        self.nums = [a, b, c]

    def order(self):
        for i in range(3):

            for k in range(3):

                if self.nums[i] <= self.nums[k]:

                    temp = self.nums[k]
                    self.nums[k] = self.nums[i]
                    self.nums[i] = temp

        return self.nums

    def dif(self):
        self.order()
        return self.nums[1]-self.nums[0] == self.nums[2]-self.nums[1]
```

```
test1 = ABC(2, 4, 6)
test2 = ABC(4, 6, 2)
test3 = ABC(4, 6, 3)

print("ABC(2, 4, 6) -->", test1.dif())
print("ABC(4, 6, 2) -->", test2.dif())
print("ABC(4, 6, 3) -->", test3.dif())
```

Quiz 18/ Fun Quiz

```
class Quiz_18():
    def __init__(self, N: str):
        self.num = N


    def calc(self):
      step = 0

      while len(self.num) > 1:
        step += 1
        prod = 1

        for i in range(len(self.num)):
          prod *= int(self.num[i])

        self.num = str(prod)

      return step

    def calc10(self):
      step = 0

      while len(self.num) > 1:
        step += 1
        prod = 1

        for i in range(len(self.num)):
          if '0' not in str(prod):
            prod *= int(self.num[i])
          else:
            break
        self.num = str(prod)

      return step

    def find10(self):
      i = 0
      best = 0
      while True:
        i+= 1
        self.num = str(i)
        if '0' not in self.num and '1' not in self.num and '2' and '5' not in self.num and '4' and '5' not in self.num and '6' and '5' not in self.num and '8' and '5' not in self.num:
          ret = self.calc10()
          if ret >= best:
            best = ret
            print(i, "steps: ", best)

          if ret == 10:
            return "The number is ", i
            break
```

```
test = Quiz_18("517")
test_1 = Quiz_18("127")
print(test.calc())
print(test_1.calc())
print(test.find10())
```

```
class Quiz():
    def  __init__(self, inp):
        self.bin = inp

    def split(self):
        spl = self.bin.split("!")
        return spl

    def convert(self):
        nums = self.split()
        tr = [4, 2, 1]
        ex = ""

        for item in nums:
            out = 0
            it = 0

            for i in range(2,-1,-1):
                mt = 2
                mx = i

                if i == 0:
                    mx = 1
                    mt = 1
                out += int(item[it])*mx*mt
                it += 1
            ex += str(out)
        return ex
```

```
test_1 = Quiz("100!000!111")
test_2 = Quiz("101!010!100")
print(test_1.convert())
print(test_2.convert())
```
