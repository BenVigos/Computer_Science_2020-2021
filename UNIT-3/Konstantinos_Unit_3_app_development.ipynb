# Bicycles class with all its functions

### Below I have my bicycles class with all it's atributes and methods. The most important of those is calc_cost which calculates the cost of the pair for that bicycle depending on the parts we need to repair and what type of bicycle it is.



```
class bicycles(wheeled):
  #constructor: This is where we define the attributes
  n_bikes = 0

  def __init__(bike, maker, model, wheelsize, biketype, parts=None):
    bike.maker = maker
    bike.model = model
    bike.wheel_size = wheelsize
    bike.bike_type = biketype
    bike.parts = [parts]
    bike.cost = 0
    bike.info= [maker, model, wheelsize, biketype, parts]
    bicycles.n_bikes+=1
    wheeled.__init__(2, "bike")




  def get_parts(bike):
    return "parts needed", bike.parts
  



  def get_bikes(bike):
    return bicycles.n_bikes




  def get_info(bike):
    return bike.info




  def set_parts(bike, part):
    #temp = input("add parts needed to repair (a, b, c, ...): ")
    #temp = temp.split(', ')
    #for i, x in enumerate(temp):
    #  bike.parts.append(temp[i])
    bike.parts.append(part)
    



  def calc_cost(bike):
    profit = 0
    if bike.bike_type == "MTB": 
      profit = 1.5

    elif bike.bike_type == "road":
      profit = 2

    else:
      profit = 1.25

    if "chain" in bike.parts:
      bike.cost += 50*profit
    
    bike.parts.append(bike.cost)
    return bike.cost

```

# Defining the owners class

### Here I'm defining the owners class with all it's attributes and methods. In this class we have the ability to create an owner, asign bicycles to that owner and calculate the price they have to pay for all the bicycles they have.

```
class owners():

  def __init__(owner, name, email, phone_n=None):
    owner.name = name
    owner.email = email
    owner.phone_n = phone_n
    owner.price = 0
    owner.bicycles = [] 




  def get_info(owner):
    owner.info.append(owner.bicycles)
    return owner.info


  def drop_bike(owner, maker, model, wheelsize, biketype, parts):
    new_bike = bicycles(maker, model, wheelsize, biketype, parts)
    owner.bicycles.append(new_bike)

  def calc_price(owner):
    tax=1.25
    price=0

    print(owner.bicycles)

    for bike in owner.bicycles:
      price += bike.calc_cost()
      tax-=0.025
    price=price*tax

    owner.info.append(price)
    return "the price is ", price
```
