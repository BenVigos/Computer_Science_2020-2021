2. Write a Python class to convert an integer to a roman numeral.

```
class RomanN():
  def __init__(self, number):
    self.number :int = number

  def convert(self):
    n = self.number
    rn = ""

    if n/1000 >= 1:
        while n/1000 >= 1:
          rn += "M"
          n -= 1000

    if n/900 >= 1:
        rn += "CM"
        n -= 900

    if n/500 >= 1:
        while n/500 >= 1:
            rn += "D"
            n -= 500

    if n/400 >= 1:
        rn += "CD"
        n -= 400

    if n/100 >= 1:
        while n/100 >= 1:
            rn += "C"
            n -= 100

    if n/90 >= 1:
        rn += "XC"
        n -= 90

    if n/50 >= 1:
        while n/50 >= 1:
            rn += "L"
            n -= 50

    if n/40 >= 1:
        rn += "XL"
        n -= 40

    if n/10 >= 1:
        while n/10 >= 1:
            rn += "X"
            n -= 10

    if n == 9:
        rn += "IX"
        n -= 9

    if n == 4:
        rn += "IV"
        n -= 4

    if n/5 >= 1:
        while n / 5 >= 1:
            rn += "V"
            n -= 5



    if n/1 >= 1:
        while n / 1 >= 1:
            rn += "I"
            n -= 1

    return rn, n




num = RomanN(3499)
print(num.convert())
```

>* Describe with words the program below.
>* Create the UML class diagram for the program below.

```
class BankAccount:
    def __init__(self):
        self.balance = 0

    def withdraw(self, amount):
        self.balance -= amount
        return self.balance

    def deposit(self, amount):
        self.balance += amount
        return self.balance

class MinimumBalanceAccount(BankAccount):
    def __init__(self, minimum_balance):
        BankAccount.__init__(self)
        self.minimum_balance = minimum_balance

    def withdraw(self, amount):
        if self.balance - amount < self.minimum_balance:
            print('Sorry, minimum balance must be maintained.')
        else:
            BankAccount.withdraw(self, amount)
```

  This program creates two classes, the class BankAccount and the class MinimumBalanceAccount which is a child class of the first one. 

  The first class has two main functions, depositing money to the account by adding the inputed amount to an attribute of the class or withdrawing money by subtracting from that attribute.

  The second class creates an object of the BankAccount class to be able to use its attributes and methods as its own. It sets a value as the minimum balance and doesn't allow you to withdraw money if that will drop the balance below the minimum.
